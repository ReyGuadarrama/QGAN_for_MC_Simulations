## **Classical Monte Carlo simulations test**

```python
from python_package.simulations import MonteCarlo
import matplotlib.pyplot as plt
```

```python
# Rolling dice simulation

sum_dice = MonteCarlo.rolling_dice(n_samples=2000)


# Visualize the probability distribution of the simulated data
plt.hist(sum_dice, bins=16, align='left', range=[0, 16], rwidth=0.8, density=True, color="cornflowerblue")
plt.xlabel("sum")
plt.ylabel("pseudo-probabilities")
plt.title("Probability Distribution")
plt.show()
```

```python
# Coin tosses simulation

coin_tosses = MonteCarlo.coin_toss(n_samples=2000, coins=10)

# Visualize the probability distribution of the simulated data
plt.hist(coin_tosses, bins=16, align='left', range=[0, 16], rwidth=0.8, density=True, color="cornflowerblue")
plt.xlabel("Number of heads")
plt.ylabel("pseudo-probabilities")
plt.title("Probability Distribution")
plt.show()
```

```python
# time decay simulation

decay_times = MonteCarlo.particle_decay(n_particles=2000, decay_constant=0.6)

# Visualize the probability distribution of the simulated data
plt.hist(decay_times, bins=16, align='left', range=[0, 16], rwidth=0.8, density=True, color="cornflowerblue")
plt.xlabel("time")
plt.ylabel("pseudo-probabilities")
plt.title("Probability Distribution")
plt.show()
```

```python
# 2D Random Walk simulation

final_positions = MonteCarlo.random_walk(n_samples=2000, n_steps=5).T

# Get the pseudo probability distribution
prob_dist = MonteCarlo.probability_distribution_2D(samples=final_positions, batch_size=2000, bins=8, lims=(-5, 5))

# Visualize the probability distribution of the simulated data
plt.figure(figsize=(8, 6))
plt.imshow(prob_dist.T, extent=[-5, 5, -5, 5], origin='lower', cmap='inferno', aspect='auto')
plt.colorbar(label='Density')
plt.title('Final positions density map')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
