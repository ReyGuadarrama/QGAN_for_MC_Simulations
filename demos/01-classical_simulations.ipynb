# **Classical Monte Carlo simulations**
---
This demo shows the usage of MonteCarlo class to generate simulations of rolling dice, coin toss, time particle decay and two dimensional random walk.

```python
from python_package.simulations import MonteCarlo as mcs
```

### **Rolling dice**
Simulation of rolling two six-side dice and sum the results 10000 times.

```python
# Rolling dice simulation
sum_dice = mcs.rolling_dice(n_samples=10000)

# Visualize the probability distribution of the simulated data
mcs.visualize_distribution_1D(samples=sum_dice, bins=16, xlabel="sum")
```

### **Coin toss**
Simulation of tossing 10 coins, 10000 times.

```python
# Coin tosses simulation
coin_tosses = mcs.coin_toss(n_samples=10000, coins=10)

# Visualize the probability distribution of the simulated data
mcs.visualize_distribution_1D(samples=coin_tosses, bins=16, xlabel="number of heads")
```

### **Particle decay time**
Simulation of 10000 particles decay time

```python
# time decay simulation
decay_times = mcs.particle_decay(n_particles=10000, decay_constant=0.6)

# Visualize the probability distribution of the simulated data
mcs.visualize_distribution_1D(samples=decay_times, bins=16, xlabel="time")
```

### **Random Walk 2D**
Simulation of the final position of 10000 samples of five steps random walk

```python
# 2D Random Walk simulation
final_positions = mcs.random_walk(n_samples=10000, n_steps=5)

# Visualize the probability distribution of the simulated data
mcs.visualize_distribution_2D(samples=final_positions, limits=(-5, 5), bins=8)
```
